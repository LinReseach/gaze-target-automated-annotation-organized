# find the value have the most frequency in every group, if the frequency of the value larger than 0.6, then take this value as result in the group, if not, set the group value as 9  

```python
df=dft[dft['case']=='visible'].copy()

l2cs_vi_agree=[]
l2cs_vi_left=[]
l2cs_invi_agree=[]
l2cs_invi_left=[]
baseline_vi_agree=[]
baseline_vi_left=[]

x_list=[0.25,0.5,0.6,0.7,0.8]
# Define a custom function to replace values in each group
for x in x_list:
    def replace_values(group):
        counts = group['l2cs'].value_counts(normalize=True)
        most_frequent_value = counts.idxmax()
       # print(counts.max())
        if counts.max() > x:

            group['l2cs'] = most_frequent_value
        else:
            group['l2cs'] = 9
        return group

    # Apply the function to each group using groupby
    df_up = df.groupby((['video','event_group'])).apply(replace_values)
    d1=df_up[(df_up['l2cs']!=9)&(df_up['l2cs']==df_up['manual'])]
    d2=df_up[(df_up['l2cs']!=9)]
    l2cs_vi_agree.append(round(len(d1)/len(d2),2))
    l2cs_vi_left.append(round(len(d2)/len(df_up),2))
    
    def replace_values(group):
        counts = group['baseline'].value_counts(normalize=True)
        most_frequent_value = counts.idxmax()
       # print(counts.max())
        if counts.max() > x:

            group['baseline'] = most_frequent_value
        else:
            group['baseline'] = 9
        return group

    # Apply the function to each group using groupby
    df_up = df.groupby((['video','event_group'])).apply(replace_values)
    d1=df_up[(df_up['baseline']!=9)&(df_up['baseline']==df_up['manual'])]
    d2=df_up[(df_up['baseline']!=9)]
    baseline_vi_agree.append(round(len(d1)/len(d2),2))
    baseline_vi_left.append(round(len(d2)/len(df_up),2))
    
    
df=dft[dft['case']=='invisible'].copy()    
for x in x_list:
    def replace_values(group):
        counts = group['l2cs'].value_counts(normalize=True)
        most_frequent_value = counts.idxmax()
       # print(counts.max())
        if counts.max() > x:

            group['l2cs'] = most_frequent_value
        else:
            group['l2cs'] = 9
        return group

    # Apply the function to each group using groupby
    df_up = df.groupby((['video','event_group'])).apply(replace_values)
    d1=df_up[(df_up['l2cs']!=9)&(df_up['l2cs']==df_up['manual'])]
    d2=df_up[(df_up['l2cs']!=9)]
    l2cs_invi_agree.append(round(len(d1)/len(d2),2))
    l2cs_invi_left.append(round(len(d2)/len(df_up),2))
    
```

## Weighted average F1-scores

```python
## from sklearn.metrics import f1_score
import numpy as np
from sklearn.metrics import f1_score


l2cs_perc=[]


for f in df[df['case']=='visible']['video'].unique():
    d=df[df['video']==f]
    human_annotations=list(d['manual'])
    print(len(d['manual']))
    model_predictions = list(d['l2cs_most_frequent'])
    print(len(d['l2cs_most_frequent']))
    f1_scores_human = f1_score(human_annotations, model_predictions, average=None)

    # Compute class weights
    class_counts = np.bincount(human_annotations)
    class_weights = class_counts / np.sum(class_counts)

    # Compute Weighted Average F1-score
    weighted_average_f1_human = np.sum(f1_scores_human * class_weights)

    l2cs_perc.append(weighted_average_f1_human)

#     if  base_perc_l<10 or l2cs_perc_l<10:        
#         print(f,base_perc_l,l2cs_perc_l,l2cs_base_l)

   
plt.plot(l2cs_perc,'bo--',label = 'Gaze_BB')

print(sum(l2cs_perc)/45)
```

```python
# -*- coding: utf-8 -*-
# =============================================================================
# =============================================================================
# DETERMINE GAZE USING FIXATION DETECTION WITH I2MC ALGORITHM
# by Erik van Haeringen
# Adapted from example by Jonathan van Leeuwen and Diederick Niehorster


import os
import sys
import pandas as pd
import numpy as np
import import_funcs as imp
import I2MC
import matplotlib.pyplot as plt
import time
from tqdm import tqdm

# =============================================================================
# SCRIPT VARIABLES
# =============================================================================
inputDir = "../input"                       # folder with input data
figDir = "../figures/fixations"             # output folder figures
procDir = "../processed"                    # output folder dataframe with fixations
plotData = False                            # plot fixation detection for each trial as png

# =============================================================================
# I2MC VARIABLES
# =============================================================================
opt = {}
# General variables for eye-tracking data
opt['xres']         = 2560.0                # maximum value of horizontal resolution in pixels
opt['yres']         = 1440.0                # maximum value of vertical resolution in pixels
opt['missingx']     = -opt['xres']          # missing value for horizontal position in eye-tracking data (example data uses -xres). used throughout the algorithm as signal for data loss
opt['missingy']     = -opt['yres']          # missing value for vertical position in eye-tracking data (example data uses -yres). used throughout algorithm as signal for data loss
opt['freq']         = 60.0                  # sampling frequency of data (check that this value matches with values actually obtained from measurement!)
# Variables for the calculation of angular measures
# These values are used to calculate noise measures (RMS and BCEA) of
# fixations. The may be left as is, but don't use the noise measures then.
# If either or both are empty, the noise measures are provided in pixels
# instead of degrees.
opt['scrSz']        = [50.9174, 28.6411]    # screen size in cm
opt['disttoscreen'] = 65.0                  # distance to screen in cm.
# # Steffen interpolation
opt['windowtimeInterp']     = 0.1                           # max duration (s) of missing values for interpolation to occur
opt['edgeSampInterp']       = 2                             # amount of data (number of samples) at edges needed for interpolation
opt['maxdisp']              = opt['xres']*0.2*np.sqrt(2)    # maximum displacement during missing for interpolation to be possible
# # K-means clustering
opt['windowtime']           = 0.2                           # time window (s) over which to calculate 2-means clustering (choose value so that max. 1 saccade can occur)
opt['steptime']             = 0.02                          # time window shift (s) for each iteration. Use zero for sample by sample processing
opt['maxerrors']            = 100                           # maximum number of errors allowed in k-means clustering procedure before proceeding to next file
opt['downsamples']          = [2, 5, 10]
opt['downsampFilter']       = False                         # use chebychev filter when downsampling? Its what matlab's downsampling functions do, but could cause trouble (ringing) with the hard edges in eye-movement data
opt['chebyOrder']           = 8.                            # order of cheby1 Chebyshev downsampling filter, default is normally ok, as long as there are 25 or more samples in the window (you may have less if your data is of low sampling rate or your window is small)
# # Fixation determination
opt['cutoffstd']            = 2.0                           # number of standard deviations above mean k-means weights will be used as fixation cutoff
opt['onoffsetThresh']       = 3.0                           # number of MAD away from median fixation duration. Will be used to walk forward at fixation starts and backward at fixation ends to refine their placement and stop algorithm from eating into saccades
opt['maxMergeDist']         = 30.0                          # maximum Euclidean distance in pixels between fixations for merging
opt['maxMergeTime']         = 30.0                          # maximum time in ms between fixations for merging
opt['minFixDur']            = 40.0                          # minimum fixation duration after merging, fixations with shorter duration are removed from output

def main():
  if not os.path.isdir(figDir):
    os.mkdir(figDir)
  if not os.path.isdir(procDir):
    os.mkdir(procDir)

  df = pd.read_csv(os.path.join(procDir, 'selectedGazeClips.csv'))
  df['trackId'] = df['session'].apply(str) + df['question'].apply(str) + df['participant'].apply(str)

  ndf = pd.DataFrame()
  ndf['trackId'] = df['trackId']
  ndf['session'] = df['session']
  ndf['condition'] = df['condition']
  ndf['question'] = df['question']
  ndf['participant'] = df['participant']
  ndf['step'] = df['step']
  ndf['x'] = df['x']
  ndf['y'] = df['y']
  ndf['time'] =  df['time'].apply(getMilliSeconds)
  ndf['absTime'] =  df['time']
  ndf['L_X'] = df['leftAbsX']
  ndf['L_Y'] = df['leftAbsY']
  ndf['R_X'] = df['rightAbsX']
  ndf['R_Y'] = df['rightAbsY']
  ndf['fixation'] = np.nan
  ndf['xPos'] = np.nan
  ndf['yPos'] = np.nan

  # Convert to relative times
  recordings = df['trackId'].unique()
  for i in tqdm(recordings, desc = "Calculating times"):
    startTime = ndf.loc[ndf['trackId'] == i,'time'].iloc[0]
    ndf.loc[ndf['trackId'] == i, 'time'] -= startTime

  # Apply I2MC algorithm
  for i in tqdm(recordings, desc = "Calculating fixations"):
    tMask = ndf['trackId'].values == i
    fix,_,_ = I2MC.I2MC(ndf.loc[tMask], opt, False, logging_offset = "      ")
    if not fix:
      print('    Fixation classification did not succeed with file {}'.format(file_name))
      continue
    if fix != False:
      if plotData:
        # pre-allocate name for saving file
        save_file = os.path.join(figDir, i + '.png')
        f = I2MC.plot.data_and_fixations(ndf.loc[tMask], fix, fix_as_line=True, res=[2560, 1440])
        f.savefig(save_file)
        plt.close(f)
      fix = pd.DataFrame(fix)
      for f in range(len(fix.index)):
        sMask = ndf['time'].values >= fix['startT'].iloc[f]
        eMask = ndf['time'].values <= fix['endT'].iloc[f]
        mask = tMask & sMask & eMask
        ndf.loc[mask, 'fixation'] = 1
        ndf.loc[mask, 'xPos'] = fix['xpos'].iloc[f]
        ndf.loc[mask, 'yPos'] = fix['ypos'].iloc[f]

  # Determine gaze direction
  # -2     not determined
  # -1     no fixation
  # 0      fixation on chat panel
  # 1-6    corresponds to participant number gaze is fixated on
  # 7      fixation else where on the screen
  posDat = pd.read_csv(os.path.join(inputDir, 'screenPositionOthers.csv'))
  ndf['gazeDirection'] = -2
  for index, row in tqdm(ndf.iterrows(), desc = "Calculating gaze", total=ndf.shape[0]):
    if row.loc['fixation'] == 1:
      ndf.loc[index, 'gazeDirection'] = 7
      mask = (posDat['session'] == row['session']) & \
        (posDat['condition'] == row['condition']) & \
        (posDat['participant'] == row['participant'])
      onScreen = posDat.loc[mask, :]
##      print(onScreen)
      for posIndex, posRow in onScreen.iterrows():
        if coordInsideRect(row['xPos'], row['yPos'], posRow['x1'], posRow['y1'], \
          posRow['width'], posRow['height']):
          ndf.loc[index, 'gazeDirection'] = posRow['other']
          break
    else:
      ndf.loc[index, 'gazeDirection'] = -1

  # Write data to file
  ndf.to_csv(os.path.join(procDir, "gazeDirection.csv"), sep=',', index=False)
  print("Done! Exported gaze data to", os.path.join(procDir, "gazeDirection.csv"))

def getMilliSeconds(value):
    """Get seconds from time."""
    h, m, s = value.split(':')
    s, ms = s.split('.')
    return float(h) * 3600000 + float(m) * 60000 + float(s) * 1000 + float(ms)

def coordInsideRect(x, y, rX1, rY1, rW, rH):
  return (x >= rX1) & (x <= (rX1 + rW)) & (y >= rY1) & (y <= (rY1 + rH))

if __name__ == '__main__':
  main()
```
